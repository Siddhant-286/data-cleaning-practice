```python
# working with Text Data

# we can also use like
'''
df['full_name] = df['full_name'].str.strip().str.title()
df['full_name'] = df['full_name'].str.strip().str.capitalize()

'''
new_df['full_name'] = new_df['full_name'].str.strip()
new_name = []
for name in new_df['full_name']:
    if isinstance(name, str) and len(name) > 0:
        name = name.lower()
        formatted_name = name[0].upper() + name[1:]
        new_name.append(formatted_name)
    else:
        new_name.append(name)
new_df['full_name'] = new_name

new_df['full_name'] = new_df['full_name'].str.rstrip('.')
print(new_df['full_name'])
```

```python
new_df['gender'] = new_df['gender'].str.strip().str.title()
new_df[new_df['gender'].isnull()]
new_df['gender'] = new_df['gender'].fillna('Unknown')
new_df['gender']
new_active = []
for i in new_df['gender']:
    if i ==  'M':
        i = 'Male'
        new_active.append(i)
    elif i == 'F':
        i = 'Female'
        new_active.append(i)
    else:
        new_active.append(i)
new_df['gender'] = new_active
new_df['gender'].unique()
```

```python
new_df['country'] = new_df['country'].str.strip('.').str.lower()

country_map = {
    'us': 'usa', 'united states': 'usa', 'usa': 'usa',
    'uk': 'united kingdom', 'u.k': 'united kingdom', 'u.k.': 'united kingdom',
    'united kingdom': 'united kingdom',
    'in': 'india', 'india': 'india',
    'fr': 'france', 'france': 'france',
}
new_df['country'] = new_df['country'].map(country_map).fillna(new_df['country'])
new_df['country']  = new_df['country'].fillna('Not Provided')
new_df['country'] = new_df['country'].str.title()
new_df['country'].unique()
```

```python
new_df['department'] = new_df['department'].str.strip('.').str.title()
new_df['department'] = new_df['department'].fillna('Not Provided')
new_df['department'] = new_df['department'].str.replace('I.T','Information Technology', regex=True)
new_df['department'] = new_df['department'].replace({'Hr':'Human Resource', 'It':'Information Technology'}, regex=False)
new_df['department'].unique()
```

```python
new_df['id'].head(20)
new_df = new_df.sort_values(by='id', ascending=True)
new_df = new_df.reset_index(drop=True)

## fixing the row number 
new_df.drop('row_num', axis = 1, inplace = True)
```

```python
new_df['age'] = pd.to_numeric(new_df['age'], errors='coerce')
new_df['age'] = new_df['age'].abs()
new_df.loc[new_df['age'] >= 100 , 'age'] = np.nan
missing_count =new_df['age'].isnull().sum()
random_values = np.random.randint(20, 60, size = missing_count)
new_df.loc[new_df['age'].isnull(), 'age'] = random_values

new_df['age'] = new_df['age'].astype(int)
```

```python
#new_df['salary'] = new_df['salary'].str.replace('k', '000', regex=True)
#new_df['salary'] = new_df['salary'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
missing_salary = new_df['salary'].isnull().sum()
random_value = np.random.randint(5000, 35000, size = missing_salary)
new_df.loc[new_df['salary'].isnull(), 'salary'] = random_value
new_df['salary'] = pd.to_numeric(new_df['salary'], errors='coerce')
new_df['salary'].value_counts()

new_df['salary'] = new_df['salary'].astype(float)
```

```python
new_df['years_experience'] = new_df['years_experience'].abs()
# wrong data fixing
new_df.loc[new_df['years_experience'] > 6, 'years_experience'] = np.nan

missing_years= new_df['years_experience'].isnull().sum()
random_value = np.random.randint(1, 5, size = missing_years)
new_df.loc[new_df['years_experience'].isnull(), 'years_experience'] = random_value

"""
count_v = len(new_df[new_df['years_experience'] == 3.0])
random_value = np.random.randint(1, 5, size = count_v)
new_df.loc[(new_df['years_experience'] == 3), 'years_experience'] = random_value 

"""
new_df['years_experience'].value_counts()
```

```python
new_df['is_active'] = new_df['is_active'].str.title()
# new_df['is_active'] = np.where(new_df['is_active'] =='TRUE' ,'Yes','No')
new_active = []
for i in new_df['is_active']:
    if i in  ['True','1','Y']:
        i = 'Yes'
        new_active.append(i)
    elif i in ['False', 'N', '0']:
        i = 'No'
        new_active.append(i)
    else:
        new_active.append(i)

new_df['is_active'] = new_active

new_df['is_active'] =  new_df['is_active'].fillna('Unknown')
```

```python
new_df.loc[new_df['performance_rating']== 'five', 'performance_rating'] = 5
new_df['performance_rating'] = pd.to_numeric(new_df['performance_rating'], errors='coerce')
new_df['performance_rating'] = new_df['performance_rating'].abs()
new_df['performance_rating'].values
new_df.loc[new_df['performance_rating'] > 6, 'performance_rating'] = np.nan
rating_mean = new_df['performance_rating'].median()

new_df['performance_rating'] = new_df['performance_rating'].fillna(rating_mean).astype(int)
```

```python
new_df['join_date'].apply(type).value_counts()
new_df['join_date'] = new_df['join_date'].replace(['not available', 'Not Provided', 'N/A'], np.nan)
new_df['join_date'].unique()
```

```python
formats = [
    '%Y-%m-%d %H:%M:%S',
    '%Y-%m-%d',           
    '%Y/%m/%d',        
    '%d-%m-%Y',          
    '%B %d, %Y',      
]

parsed = pd.Series(pd.NaT, index=new_df.index)

for fmt in formats:
    still_unparsed = parsed.isna() & new_df['join_date'].notna()
    parsed.loc[still_unparsed] = pd.to_datetime(
        new_df.loc[still_unparsed, 'join_date'], format=fmt, errors='coerce'
    )

new_df['join_date'] = parsed

print(new_df['join_date'].dtype)                 
print(new_df['join_date'].isnull().sum())      
print(new_df[new_df['join_date'].isnull()])          
```
