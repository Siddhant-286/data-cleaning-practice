```python
text_df['id'].unique()
text_df.sort_values(by = 'id', ascending=True, inplace=True)
text_df.info()
```

```python

def fix_name_order(name):
    if ',' in str(name):
        last, first = [p.strip() for p in name.split(',', 1)]
        return f"{first} {last}"
    return name

text_df['full_name'] = text_df['full_name'].astype(str).apply(fix_name_order)
```

```python
# Cleaning and Fix full_name 
text_df['full_name'] = text_df['full_name'].str.strip().str.replace(r'[,.-]',' ', regex=True).str.replace(r'\s+', ' ', regex=True)
text_df['full_name']= text_df['full_name'].str.title()
text_df['full_name'].unique()
```

```python
text_df['email'] = text_df['email'].str.lower()
text_df['email'] = (
    text_df['email'].str.replace(r'[.-]','',regex=True)
    .replace('invalidemail', np.nan)
    .str.replace(r'dotcom$', '.com', regex=True)
    .str.replace(r'at(gmail|yahoo|outlook|company)', r'@\1', regex=True)
    .str.replace(r'(gmail|yahoo|outlook)com$', r'\1.com', regex=True).str.strip()
    .str.replace(r'companycoin$', 'company.co.in', regex=True)
    .str.replace(r'(gmail|yahoo|outlook)$', r'\1.com', regex=True)
    .str.replace('DOT', '.', regex=False).str.replace('AT', '@', regex=False)
)
text_df['email'].unique()
```

```python
text_df['city'] = text_df['city'].str.title().str.strip()
text_df['city'] = text_df['city'].str.replace('Bengaluru','Bangalore', regex=True)
text_df['city'] = text_df['city'].replace({'Nyc': 'New York','Delhi' :'New Delhi' })
text_df['city'].unique()
```

```python
text_df['country'] = text_df['country'].str.strip().str.title()
new_list = {
'Uk' : 'United Kingdom',
'U.K.' :'United Kingdom', 
'In' : 'India',
'U.S.A' : 'United States Of America',
'Usa' :  'United States Of America',
'Us' :  'United States Of America'
}

text_df['country'] = text_df['country'].replace(new_list, regex=False)
text_df['country'].unique()
```

```python
text_df['gender'] = text_df['gender'].str.title().str.strip()
edit_list = {
    'F' : 'Female',
    'M' : 'Male'
}
text_df['gender'] = text_df['gender'].replace(edit_list, regex=False)
```

```python
text_df['address'] = text_df['address'].str.title().str.strip().str.rstrip(',. ')
text_df['address'] = text_df['address'].str.title().str.replace(r'\bMg\b', 'MG', regex=True)
```

```python
text_df['job_title'] = text_df['job_title'].str.title()
text_df['job_title'] = text_df['job_title'].str.replace(r'\s+',' ', regex=True)
edit_list = {
   'Sr. Software Engineer' :  'Sr Software Engineer',
   'Software Engineer' : 'Sr Software Engineer',
   'Sr Sr Software Engineer' : 'Sr Software Engineer',
}

text_df['job_title'] = text_df['job_title'].replace(edit_list, regex=False)
text_df['job_title'].unique()
```

```python
text_df['department'] = text_df['department'].str.title()
text_df['department'] = text_df['department'].replace({'Hr': 'Human Resources'})
text_df['department'].unique()
```

```python
text_df['company_name'] = text_df['company_name'].str.title().str.strip('.').str.replace('&','And', regex=False)
text_df['company_name'] = text_df['company_name'].replace('Global Traders Pvt. Ltd', 'Global Traders Pvt Ltd', regex=False)

text_df['company_name'].unique()
```

```python
text_df['product_category'] = text_df['product_category'].str.title().str.strip('.').str.replace('&','And', regex=False)
text_df['product_category'] = text_df['product_category'].str.replace('Homeandkitchen' , 'Home And Kitchen', regex=False)
text_df['product_category'].unique()
```

```python
text_df['feedback_text'] = (
    text_df['feedback_text'].str.replace(r'[!,-.:(]' , ' ', regex=True )
    .str.replace(r'\s+' ,' ', regex=True)
    .str.replace('&amp;', '&', regex=False)
)
text_df['feedback_text'] = text_df['feedback_text'].str.title().str.strip()
```

```python
text_df['status'] = text_df['status'].str.title().str.strip()
edit_list = {
    'A' : 'Active',
    'I' : 'Inactive'
}
text_df['status'] = text_df['status'].replace(edit_list, regex=False)
```

```python
text_df['notes'] = text_df['notes'].str.replace(r'[!,-.]' , ' ', regex=True ).str.replace(r'\s+',' ', regex=True)
text_df['notes'] = text_df['notes'].str.title().str.strip()
```

```python
text_df['phone_number'] = text_df['phone_number'].str.strip()
text_df['phone_number'] = (
    text_df['phone_number'].str.replace(r'[,-.()]' , '', regex=True )
    .str.replace(r'\s+', '', regex=True)
    .replace('unknown',np.nan).str.lstrip('0')
    .str.replace(r'^\+91', '', regex=True)
    .str.replace(r'\D', '', regex=True)
)

text_df['phone_number'] = text_df['phone_number'].where(text_df['phone_number'].str.len() == 10, np.nan)
```

```python
text_df.drop_duplicates(inplace=True)
text_df.reset_index(drop=True, inplace=True)
```
