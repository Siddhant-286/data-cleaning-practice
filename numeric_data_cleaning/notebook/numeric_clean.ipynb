```python
import pandas as pd
import numpy as np
```

```python
num_df['age'] = num_df['age'].str.replace(r'[-]','',regex=True)
```

```python
num_df['heart_rate'] = num_df['heart_rate'].replace(-70,70, regex=True)
num_df.loc[num_df['heart_rate'] > 180, 'heart_rate'] = np.nan
```
